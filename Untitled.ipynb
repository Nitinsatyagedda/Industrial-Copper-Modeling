```python
import pandas as pd
import numpy as np
```

```python
def load_data():
    df=pd.read_excel('COPPER_SET.xlsx')
    # Convert rubbish values in 'Material_Reference' to null
    df['material_ref'] = df['material_ref'].apply(lambda x: np.nan if str(x).startswith('00000') else x)
    # Drop the 'INDEX' column
    df.drop(columns=['id','material_ref','quantity tons'], inplace=True)
    return df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def preprocess_data(df):
    # Handle missing values with median
    df=df.dropna()
    
    # Encode categorical variables using LabelEncoder
    cat_cols = ['status','item type']
    le = LabelEncoder()
    for col in cat_cols:
        df[col] = le.fit_transform(df[col])
    return df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def before(df):
    plt.figure(figsize=(12, 5))
    plt.subplot(1,2,1)
    sns.boxplot(data=df, orient="v")
    plt.title("Boxplot (Before Treatment)")

    plt.subplot(1,2,2)
    sns.violinplot(data=df, orient="v", inner="quartile")
    plt.title("Violinplot (Before Treatment)")
```

```python
from scipy.stats import skew
```

```python
def skew_and_outlier(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Identify outliers
    outliers = (df < lower_bound) | (df > upper_bound)
    # Remove outliers
    df = df[~outliers.any(axis=1)]

    skewed_feats = df.apply(lambda x: skew(x)).sort_values(ascending=False)
    skewed_feats = skewed_feats[skewed_feats > 0.5]
    df[skewed_feats.index] = np.log1p(df[skewed_feats.index])
    return df
```

```python
def after(df):
    plt.figure(figsize=(12, 5))

    plt.subplot(1,2,1)
    sns.boxplot(data=df)
    plt.title("Boxplot (After Treatment)")

    plt.subplot(1,2,2)
    sns.violinplot(data=df)
    plt.title("Violinplot (After Treatment)")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def build_model(df, task):
    # Split dataset into features and target variable
    X = df.drop(columns=['selling_price', 'status'])
    y_test_reg = df['selling_price']
    y_test_cls = df['status']

    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test_reg = train_test_split(X, y_test_reg, test_size=0.2, random_state=42)
    X_train, X_test, y_train,y_test_cls = train_test_split(X, y_test_cls, test_size=0.2, random_state=42)
    # Regression model
    if task == "Regression":
        model_reg = LinearRegression()
        model_reg.fit(X_train, y_train)
        #y_pred_reg = model_reg.predict(X_test)
        return model_reg, X_test, y_test_reg

    # Classification model
    elif task == "Classification":
        model_cls = RandomForestClassifier(random_state=42)
        model_cls.fit(X_train, y_train)
        #y_pred_cls = model_cls.predict(X_test)
        return model_cls, X_test, y_test_cls
```

```python
import streamlit as st
```

```python
def main():
    st.title("Copper Industry ML Model")

    # Load data
    df = load_data()

    # Preprocess data
    df = preprocess_data(df)

    # EDA
    before(df)
    df=skew_and_outlier(df)
    after(df)
    

    # Model Building and Evaluation
    task = st.selectbox("Task", ("Regression", "Classification"))
    model, X_test,y_test_cls = build_model(df, task)

    # Streamlit Page for Prediction
    st.subheader("Prediction")

    if task == "Regression":
        input_values = {}
        for col in df.columns:
            if col != 'selling_price':
                input_values[col] = st.number_input(f"Enter value for {col}", value=0.0)

        if st.button("Predict"):
            input_data = pd.DataFrame([input_values])
            prediction = model.predict(input_data)
            st.write(f"Predicted Selling Price: {prediction}")

    elif task == "Classification":
        input_values = {}
        for col in df.columns:
            if col != 'status':
                input_values[col] = st.number_input(f"Enter value for {col}", value=0.0)

        if st.button("Predict"):
            input_data = pd.DataFrame([input_values])
            prediction = model.predict(input_data)
            if prediction[0] == 1:
                st.write("Predicted Status: WON")
            else:
                st.write("Predicted Status: LOST")
            st.write(f"Actual Status: {y_test_cls.iloc[0]}")
```

```python
if __name__ == "__main__":
    main()
```
